```python
import pandas as pd

# Chargement du fichier CSV
df = pd.read_csv("weather_data.csv")

# Suppression des colonnes "col1" et "col2"
df = df.drop(["astronomy", "maxtempC", "mintempC", "mintempF","maxtempF", "avgtempF", "totalSnow_cm", "sunHour", "uvIndex", "hourly"], axis=1)

# Renommer la colonne
df = df.rename(columns={"avgtempC": "temperature_moyenne_°C"})

# Enregistrement du DataFrame dans un nouveau fichier CSV
df.to_csv("temp_paris.csv", index=False)
```
